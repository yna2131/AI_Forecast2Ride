# Data Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
myData = pd.read_csv('SeoulBikeData.csv')
data_oneshot = myData.copy()
```

```python
hotEnc = OneHotEncoder()

features = data_oneshot[['Seasons']]
encoded = hotEnc.fit_transform(features).toarray()
```

```python
labels = ['Holiday', 'Functioning Day']

for label in labels:
    labelEnc = LabelEncoder()
    labelEnc = labelEnc.fit(data_oneshot[label])
    data_oneshot[label] = labelEnc.transform(data_oneshot[label])
```

```python
data_oneshot = data_oneshot.drop(columns='Seasons')

hotEnc.get_feature_names_out()

encoded_df = pd.DataFrame(encoded)
encoded_df.columns = hotEnc.get_feature_names_out()

data_oneshot = pd.concat([data_oneshot, encoded_df], axis=1)


data_oneshot.to_csv("SeoulBikeData_Cleaned.csv", index=False)
data_oneshot.head(10)
```

# Correlations Between Features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cleaned_data = pd.read_csv("SeoulBikeData_Cleaned.csv")
cleaned_data = cleaned_data.drop(columns="Date", errors="ignore")

corr = cleaned_data.corr(numeric_only=True)

print("\nCorrelation with 'Rented Bike Count':\n")
print(corr["Rented Bike Count"].sort_values(ascending=False))

plt.figure(figsize=(14, 10))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    center=0,
    vmin=-1,
    vmax=1,
    square=True,
    linewidths=0.5,
)
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.tight_layout()
plt.show()
```
